```python
from scipy import stats
import numpy as np
import pandas as pd
```

```python
# statistic: -2.5298221281347035, pvalue= 0.005706018193000826
```

```python
# ho mu = 13.2
# h1 mu < 13.2
```

```python
x_bar = 12.20 # elamların ortlaması
n = 40  # eleman sayısı
sigma = 2.5 # populasyon std
mu =  13.2  # populasyon mu 
```

```python
z = (x_bar - mu) / (sigma/np.sqrt(n))
z
```

```python
p_value = stats.norm.cdf(z)
p_value
```

```python
alpha = 0.01

if p_value<alpha:
    print('At {} level of significance, we can reject the null hypothesis in favor of alternative hypothesis.'.format(alpha))
else:
    print('At {} level of significance, we fail to reject the null hypothesis.'.format(alpha))
```

```python
soil = pd.read_excel("soil.xlsx")
soil = soil.dropna()
soil
```

```python
# H0 : mu1 = mu2
# H1 : mu1!= mu2
```

```python
soil.Soil1.mean()
```

```python
soil.Soil2.mean()
```

```python
# aradakı fark anlamlı mı 
```

```python
indTest = stats.ttest_ind(soil.Soil1,soil.Soil2)
```

```python
indTest.statistic
```

```python
indTest.pvalue
```

```python
alpha = 0.01

if indTest.pvalue < alpha:
    print("Reject the Null")
else:
    print("Fail to reject")
```

```python
weight=pd.read_excel("weight.xlsx")
weight
```

```python
weight.columns
```

```python
drop_list = ['Unnamed: 3', 'Unnamed: 4', 'Unnamed: 5',
       'Unnamed: 6', 'Unnamed: 7', 'Unnamed: 8', 'Unnamed: 9', 'Unnamed: 10',
       'Unnamed: 11', 'Unnamed: 12', 'Unnamed: 13', 'Unnamed: 14',
       'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18']
```

```python
weight = weight.drop(drop_list,axis=1)
weight = weight.dropna()
weight
```

```python
weight.drop("ID",axis=1,inplace=True)
weight["Diff"] = weight["starting"] - weight["ending"]
```

```python
weight
```

```python
#H0: dbar = 0

#H1: dbar > 0
```

```python
stats.ttest_rel(weight.starting, weight.ending, alternative="greater")
```

```python
one_tailed_paired_t = stats.ttest_rel(weight.starting, weight.ending, alternative="greater")
```

```python
alpha = 0.01

if one_tailed_paired_t.pvalue < alpha:
    print("Reject the Null")
else:
    print("Fail to reject")
```
